```python
import pandas as pd
import os
# Paths are relative to the repo root - run this notebook from there.
```

```python
df1 = pd.read_csv("./scripts/pubmed_output/healthbase_articles_train_unbalanced_55K.csv")
df2 = pd.read_csv("./scripts/pubmed_output/healthbase_articles_val_unbalanced_6.5K.csv")
df3 = pd.read_csv("./scripts/pubmed_output/healthbase_articles_test_unbalanced_4.5K.csv")

df = pd.concat([df1, df2, df3], ignore_index=True)
```

```python
1427/(df.shape[0]-1427)*100
```

```python
import matplotlib.pyplot as plt
# Check label balance
print(df["case"].value_counts())

# Set the size of the figure
fig, ax = plt.subplots(figsize=(10, 10))  

df.groupby('case').size().plot(kind='pie',
                               y="v1",
                               label="Type",
                               autopct='%1.1f%%',
                               ax=ax,
                               fontsize=14,
                               textprops={'fontsize':   28})  

plt.show()
```
