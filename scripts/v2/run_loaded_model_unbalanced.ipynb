```python
# Report the number of sentences.
print('Number of test sentences: {:,}\n'.format(df.shape[0]))

# Create sentence and label lists.
# v2 uses CI_TiAb (contraindication + intervention + title + abstract), which is
# what this model was trained on — unlike v1, which uses TiAb (title + abstract).
sentences = list(df.CI_TiAb.values)
labels = list(df.case.to_numpy())

# Tokenize all of the sentences and map the tokens to their word IDs.
input_ids = []
attention_masks = []

# For every sentence...
for sent in sentences:
    # `encode_plus` will:
    #   (1) Tokenize the sentence.
    #   (2) Prepend the `[CLS]` token to the start.
    #   (3) Append the `[SEP]` token to the end.
    #   (4) Map tokens to their IDs.
    #   (5) Pad or truncate the sentence to `max_length`
    #   (6) Create attention masks for [PAD] tokens.
    encoded_dict = tokenizer.encode_plus(
                        sent,                      # Sentence to encode.
                        add_special_tokens = True, # Add '[CLS]' and '[SEP]'
                        truncation = True,
                        max_length = 512,           # Pad & truncate all sentences.
                        pad_to_max_length = True,
                        return_attention_mask = True,   # Construct attn. masks.
                        return_tensors = 'pt',     # Return pytorch tensors.
                   )
    
    # Add the encoded sentence to the list.    
    input_ids.append(encoded_dict['input_ids'])
    
    # And its attention mask (simply differentiates padding from non-padding).
    attention_masks.append(encoded_dict['attention_mask'])

# Convert the lists into tensors.
input_ids = torch.cat(input_ids, dim=0)
attention_masks = torch.cat(attention_masks, dim=0)
labels = torch.tensor(labels)

# Set the batch size for DataLoader.  
batch_size = 64

# Create the DataLoader.
# prediction_data = TensorDataset(input_ids, attention_masks)
prediction_data = TensorDataset(input_ids, attention_masks, labels)
prediction_sampler = SequentialSampler(prediction_data)
prediction_dataloader = DataLoader(prediction_data, sampler=prediction_sampler, batch_size=batch_size)
```

```python
# # ALTERNATIVE: Group batches based on 'rapport_number'
# # Each batch processed will contain only articles for a single rapport_number.
# #
# # DISABLED: this yields batches of hundreds/thousands of articles for large
# # reports, which overflows memory and forces SSD swap (~6h vs ~2min with a
# # fixed batch size). Instead keep the fixed-batch SequentialSampler above and
# # reconstruct per-report groups after prediction (see the MCC-per-rapport cell).

# class RapportBatchSampler(torch.utils.data.Sampler):
#     def __init__(self, df):
#         grouped = df.groupby('rapport_number')
#         self.rapport_numbers = sorted(grouped.groups.keys())
#         self.batches = [grouped.groups[r].tolist() for r in self.rapport_numbers]
        
#     def __iter__(self):
#         for batch in self.batches:
#             yield batch
            
#     def __len__(self):
#         return len(self.batches)

# # Instantiate the sampler and extract the unique rapport numbers in the correct batch order
# batch_sampler = RapportBatchSampler(df)
# rapport_numbers = batch_sampler.rapport_numbers

# # Create prediction dataset and the custom DataLoader
# prediction_data = TensorDataset(input_ids, attention_masks, labels)
# prediction_dataloader = DataLoader(prediction_data, batch_sampler=batch_sampler)

# print(f"Created DataLoader with {len(prediction_dataloader)} batches (one per rapport_number).")
```

```python
# Prediction on set of articles
print('Predicting labels for {:,} found articles...'.format(len(input_ids)))

# Put model in evaluation mode
model.eval()

# Tracking variables 
predictions , true_labels = [], []
# predictions = []

# Predict 
for batch in prediction_dataloader:
  # Add batch to GPU
  batch = tuple(t.to(device) for t in batch) 
  
  # Unpack the inputs from our dataloader
  b_input_ids, b_input_mask, b_labels = batch
  # b_input_ids, b_input_mask = batch
  
  # Telling the model not to compute or store gradients, saving memory and 
  # speeding up prediction
  with torch.no_grad():
      # Forward pass, calculate logit predictions.
      result = model(b_input_ids, 
                     token_type_ids=None, 
                     attention_mask=b_input_mask,
                     return_dict=True)

  logits = result.logits

  # Move logits and labels to CPU
  logits = logits.detach().cpu().numpy()
  label_ids = b_labels.to('cpu').numpy()
  
  # Store predictions and true labels
  predictions.append(logits)
  true_labels.append(label_ids)

print('    DONE.')
```

### MCC is used as metric (is also used for imbalanced datasets)

```python
from sklearn.metrics import matthews_corrcoef

matthews_set = []

# Evaluate each test batch using Matthew's correlation coefficient
print('Calculating Matthews Corr. Coef. for each batch...')

# For each input batch...
for i in range(len(true_labels)):
  
  # The predictions for this batch are a 2-column ndarray (one column for "0" 
  # and one column for "1"). Pick the label with the highest value and turn this
  # in to a list of 0s and 1s.
  pred_labels_i = np.argmax(predictions[i], axis=1).flatten()
  
  # Calculate and store the coef for this batch.  
  matthews = matthews_corrcoef(true_labels[i], pred_labels_i)                
  matthews_set.append(matthews)
```

```python
# MCC per rapport_number, computed from the fixed-batch-size predictions.
# SequentialSampler preserves the row order of `df`, so the concatenated
# predictions align 1:1 with df rows and can be grouped by rapport_number
# after the fact — no per-report batch sampler needed.
import seaborn as sns
from sklearn.metrics import accuracy_score
from matplotlib.patches import Patch

flat_preds = np.argmax(np.concatenate(predictions, axis=0), axis=1)
flat_trues = np.concatenate(true_labels, axis=0)

assert len(flat_preds) == len(df), "predictions and df are out of sync — rerun prediction"

results = pd.DataFrame({
    'rapport_number': df['rapport_number'].values,
    'true_label': flat_trues,
    'pred_label': flat_preds,
})

# MCC is the correlation between true and predicted labels, so it needs variance
# in the true labels. A report whose articles are all one class has a constant
# truth vector -> MCC is mathematically undefined (0/0). For those we fall back
# to accuracy, which for a single-class report equals specificity (all-negative)
# or sensitivity (all-positive), and flag them so they're not read as real MCC.
def score_report(g):
    single_class = g['true_label'].nunique() == 1
    if single_class:
        value = accuracy_score(g['true_label'], g['pred_label'])
    else:
        value = matthews_corrcoef(g['true_label'], g['pred_label'])
    return pd.Series({'score': value, 'single_class': single_class})

per_report = results.groupby('rapport_number').apply(score_report)

plt.figure(figsize=(16, 12))
palette = ['tab:orange' if sc else 'tab:blue' for sc in per_report['single_class']]
sns.barplot(x=per_report.index, y=per_report['score'], palette=palette, errorbar=None)

# Legend explaining the two bar colors.
plt.legend(handles=[
    Patch(color='tab:blue', label='MCC (multi-class report)'),
    Patch(color='tab:orange', label='Accuracy fallback (single-class report, MCC undefined)'),
])
plt.title('Per-Rapport Score: MCC where defined, accuracy where not')
plt.ylabel('Score (-1 to +1)')
plt.xlabel('Rapport Number')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

n_single = int(per_report['single_class'].sum())
print(f'{n_single} of {len(per_report)} reports are single-class '
      f'(orange bars show accuracy, not MCC).')
```

## Check performance in this unbalanced set

```python
import numpy as np
from sklearn.metrics import balanced_accuracy_score, recall_score, roc_curve, roc_auc_score, f1_score
import matplotlib.pyplot as plt

# Combine the results across all batches. 
flat_predictions = np.concatenate(predictions, axis=0)

# For each sample, pick the label (0 or 1) with the higher score.
flat_predictions = np.argmax(flat_predictions, axis=1).flatten()

# Combine the correct labels for each batch into a single list.
flat_true_labels = np.concatenate(true_labels, axis=0)

# Calculate metrics
print('Sensitivity: ', recall_score(flat_true_labels, flat_predictions))
print('Specificity: ', recall_score(flat_true_labels, flat_predictions, pos_label=0))
print('Balanced accuracy: ', balanced_accuracy_score(flat_true_labels, flat_predictions))
print('F1 score: ', f1_score(flat_true_labels, flat_predictions))

# Plot ROC-curve
predictions_1 = [pred[1] for pred in np.concatenate(predictions, axis=0)]

fpr, tpr, thresholds = roc_curve(flat_true_labels, predictions_1)
roc_auc = roc_auc_score(flat_true_labels, predictions_1)

# Determine the corresponding FPR values for the desired sensitivity levels
target_sensitivities = [0.95, 0.93, 0.90, 0.85]
corresponding_fpr = []

for ts in target_sensitivities:
    # Find indices where tpr is close to the target sensitivity
    close_indices = np.where(np.isclose(tpr, ts, atol=0.01))[0]
    
    # Among those indices, choose the one with the lowest FPR (highest specificity)
    best_idx = close_indices[np.argmin(fpr[close_indices])]
    
    corresponding_fpr.append(fpr[best_idx])

# Plot the ROC curve along with these points
plt.figure(figsize=(8, 8))    
plt.plot(fpr, tpr, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')

# Plot the points for the target sensitivities
for ts, cfpr in zip(target_sensitivities, corresponding_fpr):
    plt.plot(cfpr, ts, 'o', markersize=10, label=f'Sens {ts*100:.0f}% (Spec {((1-cfpr)*100):.2f}%)')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.gca().set_aspect('equal', adjustable='box')
plt.show()

# Calculate and print accuracy for target sensitivities
for ts, cfpr in zip(target_sensitivities, corresponding_fpr):
    TP = ts * np.sum(flat_true_labels == 1)
    FP = cfpr * np.sum(flat_true_labels == 0)
    TN = (1 - cfpr) * np.sum(flat_true_labels == 0)
    FN = (1 - ts) * np.sum(flat_true_labels == 1)
    
    accuracy = (TP + TN) / (TP + FP + TN + FN)
    print(f'Accuracy for Sensitivity {ts*100:.0f}%: {accuracy:.2f}')

# Calculate and print balanced accuracy for target sensitivities
for ts, cfpr in zip(target_sensitivities, corresponding_fpr):
    specificity = 1 - cfpr
    balanced_acc = (ts + specificity) / 2
    print(f'Balanced Accuracy for Sensitivity {ts*100:.0f}%: {balanced_acc:.2f}')


# Calculate and print F1 score for target sensitivities
for ts, cfpr in zip(target_sensitivities, corresponding_fpr):
    specificity = 1 - cfpr
    TP = ts * np.sum(flat_true_labels == 1)
    FP = cfpr * np.sum(flat_true_labels == 0)
    precision = TP / (TP + FP)
    f1_score = 2 * (precision * ts) / (precision + ts)
    print(f'F1 Score for Sensitivity {ts*100:.0f}%: {f1_score:.2f}')
```

```python
## AUROC with 95% CI (DeLong)
import sys
import numpy as np

# Absolute path to this project's scripts/ dir, so the import works regardless of
# the notebook's current working directory.
PROJECT_SCRIPTS = './scripts'
sys.path.append(PROJECT_SCRIPTS)
from delong import delong_ci

# Positive-class score per article, aligned to flat_true_labels.
# AUROC is rank-based, so raw logits and probabilities give the same value.
scores = np.concatenate(predictions, axis=0)[:, 1]
y_true = np.concatenate(true_labels, axis=0)

auc, (lo, hi) = delong_ci(y_true, scores, alpha=0.95)
print(f"V2 (CI_TiAb) AUROC = {auc:.3f} (95% CI {lo:.3f}-{hi:.3f})")
print(f"n = {len(y_true)}  |  positives = {int(y_true.sum())}  |  negatives = {int((y_true == 0).sum())}")
```

```python
## AUPRC (average precision) with 95% bootstrap CI
from sklearn.metrics import average_precision_score

scores = np.concatenate(predictions, axis=0)[:, 1]   # positive-class logit
y_true = np.concatenate(true_labels, axis=0)

# AUPRC is rank-based, so raw logits and probabilities give the same value.
auprc = average_precision_score(y_true, scores)
prevalence = y_true.mean()   # AUPRC of a random classifier == positive prevalence

# average_precision has no simple closed-form SE, so use a stratified bootstrap
# (resample positives and negatives separately so every replicate keeps both
# classes and a stable prevalence).
rng = np.random.default_rng(42)
pos_idx = np.where(y_true == 1)[0]
neg_idx = np.where(y_true == 0)[0]
boot = np.empty(2000)
for b in range(boot.size):
    idx = np.concatenate([rng.choice(pos_idx, pos_idx.size, replace=True),
                          rng.choice(neg_idx, neg_idx.size, replace=True)])
    boot[b] = average_precision_score(y_true[idx], scores[idx])
lo, hi = np.percentile(boot, [2.5, 97.5])

print(f"V2 (CI_TiAb) AUPRC = {auprc:.3f} (95% CI {lo:.3f}-{hi:.3f})")
print(f"baseline (positive prevalence) = {prevalence:.3f}  |  "
      f"n = {len(y_true)}  positives = {int(y_true.sum())}  negatives = {int((y_true == 0).sum())}")
```

```python
## F-beta (F2 and F0.5) at fixed sensitivity targets
import numpy as np
import pandas as pd
from sklearn.metrics import roc_curve

scores = np.concatenate(predictions, axis=0)[:, 1]   # positive-class logit
y_true = np.concatenate(true_labels, axis=0)
P = int((y_true == 1).sum())
N = int((y_true == 0).sum())

fpr, tpr, thresholds = roc_curve(y_true, scores)

def fbeta(precision, recall, beta):
    # F2 (beta=2) weights recall higher; F0.5 (beta=0.5) weights precision higher.
    b2 = beta ** 2
    denom = b2 * precision + recall
    return (1 + b2) * precision * recall / denom if denom > 0 else 0.0

# Same robust operating-point rule as the ROC cells: the first point with
# tpr >= target has the lowest FPR (=> highest precision) among the points that
# meet the target sensitivity. Precision and F-beta are taken from the exact
# confusion matrix at that point (TP = tpr*P, FP = fpr*N are exact counts).
target_sensitivities = [0.95, 0.93, 0.90, 0.85]
rows = []
for ts in target_sensitivities:
    meets = np.where(tpr >= ts)[0]
    if len(meets) == 0:
        print(f'Skipping sensitivity {ts*100:.0f}%: not achievable (max TPR = {tpr.max():.3f})')
        continue
    i = meets[0]
    recall = tpr[i]
    TP = recall * P
    FP = fpr[i] * N
    precision = TP / (TP + FP) if (TP + FP) > 0 else 0.0
    rows.append({
        'target_sens': f'{ts*100:.0f}%',
        'achieved_sens': round(recall, 3),
        'specificity': round(1 - fpr[i], 3),
        'precision': round(precision, 3),
        'F2': round(fbeta(precision, recall, 2), 3),
        'F0.5': round(fbeta(precision, recall, 0.5), 3),
        'threshold_logit': round(float(thresholds[i]), 3),
    })

fbeta_table = pd.DataFrame(rows)
print('V2 (CI_TiAb) - F-beta at fixed sensitivity targets '
      '(F2 favours recall, F0.5 favours precision):')
fbeta_table
```
